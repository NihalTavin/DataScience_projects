```python
import numpy as np
import pandas as pd
```

```python
movie_data = pd.read_csv('IMDb Movies India.csv', encoding="ISO-8859-1")
```

```python
movie_data.head(10)
```

```python
movie_data=movie_data.drop(columns=["Name", "Director", "Actor 1", "Actor 2", "Actor 3"])
```

```python
movie_data=movie_data.dropna()
movie_data.isnull().sum()
```

```python
movie_data= pd.get_dummies(movie_data, columns=['Genre'], drop_first=True)
```

```python
movie_data["Year"] = movie_data["Year"].str.extract(r'(\d+)').astype(int)
```

```python
movie_data["Duration"] = movie_data["Duration"].str.replace(" min", "", regex=True).astype(int)
```

```python
movie_data["Votes"] = movie_data["Votes"].str.replace(",", "").astype(int)
```

```python
X = movie_data.drop(columns=['Rating'])
y = movie_data['Rating']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
prediction = pd.DataFrame({"Actual Rating": y_test, "Predicted Rating": y_pred})
print(prediction.head(10))
```
