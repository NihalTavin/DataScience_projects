```python
import numpy as np
import pandas as pd
```

```python
car_price_data=pd.read_csv('car_purchasing.csv', encoding='ISO-8859-1')
```

```python
car_price_data.isnull().sum()
```

```python
car_price_data.head()
```

```python
car_price_data=car_price_data.drop(columns=['customer name','customer e-mail','country','gender'],axis=1)
```

```python
y = car_price_data['car purchase amount']
X = car_price_data.drop(['car purchase amount'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=4)
```

```python
print(X_train.dtypes)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(y_pred)
```

```python
predictions = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(predictions.head(10))
```

```python
from sklearn.metrics import r2_score
y_train_pred = lr.predict(X_train)
training_accuracy = r2_score(y_train, y_train_pred)
testing_accuracy = r2_score(y_test, y_pred)
print(f"R² Score on Training Data: {training_accuracy * 100:.2f}%")
print(f"R² Score on Testing Data: {testing_accuracy * 100:.2f}%")
```
